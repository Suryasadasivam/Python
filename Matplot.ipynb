```python
x=np.arange(1,11)
y=np.random.randint(-200,200,10)
z=np.random.randint(-100,150,10)
```

```python
fig=plt.figure(figsize=(5,5))# need to mention this is figure
p1=fig.add_subplot(111,projection='3d')
p1.scatter(x,y,z)
plt.show()
```

```python
fig=plt.figure(figsize=(10,10))# need to mention this is figure
p1=fig.add_subplot(111,projection='3d')
p1.scatter(x,y,z)
p1.set_xlabel("xaxis")
p1.set_ylabel("yaxis")
p1.set_zlabel("zaxis")
plt.show()
```

```python
fig=plt.figure(figsize=(5,5))# need to mention this is figure
p1=fig.add_subplot(221,projection='3d')
p1.scatter(x,y,z)
p1.set_xlabel("xaxis")
p1.set_ylabel("yaxis")
p1.set_zlabel("zaxis")
p2=fig.add_subplot(224,projection='3d')
p2.scatter(x,y,z)
p2.set_xlabel("xaxis")
p2.set_ylabel("yaxis")
p2.set_zlabel("zaxis")
plt.show()
```
